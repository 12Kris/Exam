```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Dataset.csv'
data = pd.read_csv(file_path)

average_price = data['close1'].mean()
max_price = data['close1'].max()
min_price = data['close1'].min()
median_price = data['close1'].median() 
volatility = data['close1'].std()
hourly_avg = data.groupby('hour')['close1'].mean() 

print(f"Середня ціна закриття: {average_price}")
print(f"Максимальна ціна закриття: {max_price}")
print(f"Мінімальна ціна закриття: {min_price}")
print(f"Медіана ціни закриттяж: {median_price}")
print(f"Волатильність (стандартне відхилення): {volatility}")

plt.figure(figsize=(12, 6))
plt.plot(data['close1'], label='Ціна закриття', color="pink")
plt.title('Графік ціни закриття')
plt.xlabel('Час')
plt.ylabel('Ціна закриття')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.hist(data['close1'], bins=50, color='pink')
plt.title('Гістограма розподілу цін закриття')
plt.xlabel('Ціна закриття')
plt.ylabel('Частота')
plt.show()

plt.figure(figsize=(12, 6))
hourly_avg.plot(kind='bar')
plt.title('Середня ціна закриття за годинами')
plt.xlabel('Година')
plt.ylabel('Середня ціна закриття')
plt.show()
```
